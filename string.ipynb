```python
string=input()
new_str=" "
for i in string:
  if'a'<=i<='z':
    new_str+=chr(ord(i)-32)
  elif'A'<=i<='Z':
      new_str+=chr(ord(i)+32)
  else:
    new_str+=i
print(f"changed string:{new_str}")

```
