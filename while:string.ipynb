```python
a=int(input())
b=int(input())
hcf=1
i=1
while i<=a and i<=b:
  if a%i==0 and b%i==0:
    hcf=i
  i+=1
print("hcf=",hcf,end=1" ")

```

```python
n=int(input())
binary=""
while n>0:
  remainder=n%2
  binary=str(remainder)+binary
  n=n//2
  print(binary,end=" ")
```

```python
start=int(input())
end=int(int(input()))
while start<=end:
  print(start,end=" ")
  start=start+2
```

```python
start=int(input())
end=int(input())
while start<=end:
  if start+2>end:
     print(start,end=" ")
  else:
     print(start,end="+")
  start=start+2
```

```python
name=input("enter a string:")
for i in range(len(name)):
  for j in range(len(name)):
    if i==j:
      print(name[i],end=" ")
    else:
      print("*",end=" ")
  print()
```

```python
s=(input())
n=len(s)
for i in range(n):
  for j in range(n):
    if i==j:
      print(s[i],end=" ")
    elif i+j==n-1:
      print(s[i],end=" ")
    else:
      print("*",end=" ")
  print()
```
